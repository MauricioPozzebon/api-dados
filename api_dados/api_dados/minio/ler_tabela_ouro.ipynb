```python
import s3fs
import pandas as pd

# Inicializando o sistema de arquivos MinIO
fs = s3fs.S3FileSystem(
    key='ROOT',  # Seu MINIO_ROOT_USER
    secret='PASSWORD',  # Sua MINIO_ROOT_PASSWORD
    client_kwargs={'endpoint_url': 'http://localhost:9000'}
)

# Caminho do arquivo Parquet na camada OURO
ouro_filename = 'dados-mercado/OURO/vendas_summary.parquet'

# Verificando se o arquivo Parquet existe na camada OURO
if fs.exists(ouro_filename):
    # Lendo o arquivo Parquet em um DataFrame
    with fs.open(ouro_filename, 'rb') as f:
        df = pd.read_parquet(f)
        
    # Exibindo a tabela
    print("Tabela de vendas acumuladas:")
    print(df)
else:
    print("Nenhum arquivo Parquet encontrado na camada OURO.")

df
```
